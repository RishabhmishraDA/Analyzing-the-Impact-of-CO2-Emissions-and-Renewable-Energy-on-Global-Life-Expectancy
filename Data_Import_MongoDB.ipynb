### Importing Libraries

```python
import pymongo
import pandas as pd
import json
```

### Connecting to MongoDB Server

```python
carbon_emissions_client = pymongo.MongoClient("mongodb://localhost:27017/")
```

### Loading Data for Carbon Emissions

```python
carbon_emissions_df = pd.read_csv("per-capita-co-emissions.csv")
```

```python
carbon_emissions_df.shape
```

```python
carbon_emissions_df.head()
```

```python
carbon_emissions_df.tail()
```

```python
carbon_emissions_data = carbon_emissions_df.to_dict(orient="records")
```

```python
carbon_emissions_data
```

### Importing Data into MongoDB

```python
carbon_emissions_db=carbon_emissions_client['CO2_Renew_Life_DB']
```

```python
carbon_emissions_db.carbon_emissions.insert_many(carbon_emissions_data)
```
